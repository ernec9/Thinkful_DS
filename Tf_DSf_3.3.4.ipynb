_Let us consider the following distributions (Rayleigh, Lognormal, Exponential, Geometric, Power, Zipf) below_

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

rayleigh = np.random.rayleigh(3, 10000)

# Plot a histogram.
plt.hist(rayleigh, bins=100)

plt.axvline(rayleigh.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(rayleigh.mean() + rayleigh.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(rayleigh.mean()-rayleigh.std(), color='b', linestyle='dashed', linewidth=2) 

plt.title('Rayleigh')
# Print the histogram.
plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

logn = np.random.lognormal(3 ,1, 10000)

plt.hist(logn, bins=100)

plt.axvline(logn.mean(), color='b', linestyle='solid', linewidth=2)

plt.axvline(logn.mean() + logn.std(), color='b', linestyle='dashed', linewidth=2)

plt.axvline(logn.mean()-logn.std(), color='b', linestyle='dashed', linewidth=2) 

plt.title('Lognormal')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

expon = np.random.exponential(0.5,1000)

plt.hist(expon, bins=100)

plt.axvline(expon.mean(), color='b', linestyle='solid', linewidth=2)

plt.axvline(expon.mean() + expon.std(), color='b', linestyle='dashed', linewidth=2)

plt.axvline(expon.mean()-expon.std(), color='b', linestyle='dashed', linewidth=2) 

plt.title('Exponential')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

geom = np.random.geometric(0.2,1000)


plt.hist(geom, bins=100)

plt.axvline(geom.mean(), color='b', linestyle='solid', linewidth=2)

plt.axvline(geom.mean() + geom.std(), color='b', linestyle='dashed', linewidth=2)

plt.axvline(geom.mean()-geom.std(), color='b', linestyle='dashed', linewidth=2) 

plt.title('Geometric')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

power = np.random.power(0.2,1000)

plt.hist(power, bins=100)

plt.axvline(power.mean(), color='b', linestyle='solid', linewidth=2)

plt.axvline(power.mean() + power.std(), color='b', linestyle='dashed', linewidth=2)

plt.axvline(power.mean()-power.std(), color='b', linestyle='dashed', linewidth=2) 

plt.title('Power')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

zipf = np.random.power(2,1000)

plt.hist(zipf, bins=100)

plt.axvline(zipf.mean(), color='b', linestyle='solid', linewidth=2)

plt.axvline(zipf.mean() + zipf.std(), color='b', linestyle='dashed', linewidth=2)

plt.axvline(zipf.mean()-zipf.std(), color='b', linestyle='dashed', linewidth=2) 

plt.title('Zipf')
plt.show()
```

_Among the 6 distributions above, Rayleigh seems to be the best described by mean and standard deviation, while Zipf and Lognormal are relatively
better descibed compared with Exponential, Geometric and Power distributions._

Generate two normally-distributed variables, one with a mean of 5 and standard deviation of 0.5, and the other with a mean of 10 and standard deviation of 1.

Add them together to create a third variable.

Graph the third variable using a histogram.

Compute the mean and standard deviation and plot them as vertical lines on the histogram.

Evaluate the descriptive statistics against the data.

```python
norm1=np.random.normal(5,0.5,1000)
norm2=np.random.normal(10,1,1000)

norms=norm1+norm2

plt.hist(norms, bins=100)

plt.axvline(norms.mean(), color='b', linestyle='solid', linewidth=2)

plt.axvline(norms.mean() + norms.std(), color='b', linestyle='dashed', linewidth=2)

plt.axvline(norms.mean()-norms.std(), color='b', linestyle='dashed', linewidth=2) 

```

_For the Normal distribution, the mean and standard distribution are much better descriptors compared with the other
distributions above._

_This may in part be attributed to the symmetry of the Normal distribution_
